```python
A = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'kema'], 
                  'Tezligi':['70', '180', '45']})
                  
A
```

```python
B = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'tramvay'], 
                  'Narxi':['98000', '10000', '65000']})
                  
B
```

```python
pd.merge(A, B, how='inner')
```

```python
pd.merge(A, B, how='outer')
```

```python
pd.merge(A, B, how='left')
```

```python
pd.merge(A, B, how='right')
```

```python
A = pd.DataFrame({'Transport_turi': ['poezd', 'avtomabil', 'kema'], 
                  'Tezligi':['70', '180', '45']})
                  
A
```

```python
pd.merge(A, B, left_on='Transport_turi', right_on='Transport')
```

```python
A = A.set_index('Transport_turi')
```

```python
A
```

```python
B
```

```python
pd.merge(A, B, left_index=True, right_on='Transport')
```

```python
B = B.set_index('Transport')
```
