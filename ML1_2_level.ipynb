```python
import pandas as pd
import numpy as np
import sklearn # scikit-learn kutubxonasi
```

```python
df = pd.read_csv('/content/uy_data1.csv', index_col=0)
df
```

```python
df.drop(4089, inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
housing = train_set.drop('price', axis=1)
housing_labels = train_set['price'].copy()
```

```python
housing
```

```python
housing_labels
```

```python
housing.isnull().sum()
```

```python
housing_loc = housing[['location']]
housing_dist = housing[['district']]
```

```python
housing_dist
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
housing_loc_encoded = ordinal_encoder.fit_transform(housing_loc)
housing_loc_encoded[:10]
```

```python
ordinal_encoder.categories_
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
housing_dist_1hot = cat_encoder.fit_transform(housing_dist)
housing_dist_1hot.toarray()
```

```python
housing_num = housing.drop(['location', 'district'], axis=1)
housing_num['location'] = housing_loc_encoded
housing_num
```

```python
housing_num.append(housing_dist_1hot.toarray())
housing_num
```

```python
from sklearn.preprocessing import MinMaxScaler
min_mix_scaler = MinMaxScaler()

min_mix_scaler.fit_transform(housing_num)
```

```python
from sklearn.preprocessing import StandardScaler
standart_scaler = StandardScaler()
standart_scaler.fit_transform(housing_num)
```
